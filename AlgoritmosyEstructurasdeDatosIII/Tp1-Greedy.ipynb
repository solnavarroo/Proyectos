```python
## Solana Navarro.
## LU: 906/22

## I AM VERY BUSY es un ejercico para realizar con Greedy.
## La estrategia Greedy es: Para poder maximizar la cantidad de actividades voy a elegir
## la actividad que primero termine. Esto lo hare iterativamente fijandome que
## la nueva actividad por elegir no se pise con la anteriormente elegida.


## Funcion para ordenar las tuplas mirando la segunda posicion
def segundatupla(tupla):
  return(tupla[1])


## Funcion donde recibe las actividades (ya ordenadas) y va una por una quedandose con las que no se pisan.
def dia(actividades, N):
     cantactividades = 1
     fin = actividades[0][1]
     for i in range(1, N):
       if fin <= actividades[i][0]:
          fin = actividades[i][1]
          cantactividades += 1
     return cantactividades

T = int(input())
res = []

for i in range(T):
  N = int(input())
  actividades = []
  for j in range(N):
    tiempo = input().split()
    actividades.append((int(tiempo[0]), int(tiempo[1])))

  ##ordeno segun la segunda tupla usando la funcion segundatupla
  actividades.sort(key = segundatupla)

  res.append(dia(actividades, N))


for j in range(len(res)):
  print(res[j])
```
