```python
## Solana Navarro
## 906/22

## Importo Queue para hacer una cola
from queue import Queue

## Mi idea inicial fue representar los tuneles con aristas dirigidas.
## Cada aula tiene un atajo que conecta con otra aula a traves de un tunel unidireccional.
## Luego, decidi que para representar el camino que Juli realiza sin tomar atajos por tuneles,
## era una buena idea unir las aulas con aristas bidireccionales. Esto permite que Juli pueda
## ir y volver entre aulas adyacentes sin atajos.

## Para encontrar la minima cantidad de energia, pense que usar BFS seria una buena idea,
## ya que en BFS todas las aristas tienen el mismo peso. Esto es importante porque queremos
## modelar los caminos de manera que los tuneles tengan un costo de 1 unidad de energia
## y caminar entre aulas adyacentes tenga un costo proporcional a la distancia.

## Note que si queria caminar del aula i al aula j, siendo i < j y |i - j| > 1,
## era equivalente poner una arista de peso |i - j| o una secuencia de aristas entre aulas adyacentes:
## una de i a i+1, otra de i+1 a i+2, ..., hasta llegar a j. Esto se debe a que tomar ese camino
## entre aulas consecutivas es lo mismo que una arista directa con peso |i - j|.

## Con esa idea planteada, lo que queda es correr BFS desde el aula inicial (el nodo 1)
## y calcular la distancia minima (numero de aristas) hasta cada aula (nodo), tomando en cuenta
## los atajos que permiten reducir el costo a 1 en ciertos casos.

def BFS(listadeadyacencias, v):
    cola = Queue()
    cola.put(v)
    distancia = [-1] * len(listadeadyacencias)
    distancia[v] = 0

    while(not cola.empty()):
      nodoactual = cola.get()
      for vecino in listadeadyacencias[nodoactual]:
          if distancia[vecino] == -1:
             distancia[vecino] = distancia[nodoactual] + 1
             cola.put(vecino)

    return distancia

## Input

n = int(input())
tuneles = list(map(int, input().split()))

listadeadyacencias = [[] for _ in range(n)]

for j in range(1, n-1):
    listadeadyacencias[j].append(j-1)
    listadeadyacencias[j].append(j+1)

listadeadyacencias[0].append(1)
listadeadyacencias[n-1].append(n-2)

for i in range(n):
  if (tuneles[i]-1) not in listadeadyacencias[i] and (tuneles[i]-1)!= i:
      listadeadyacencias[i].append(tuneles[i]-1)

distancias = BFS(listadeadyacencias, 0)
print(' '.join(map(str, distancias)))
```
