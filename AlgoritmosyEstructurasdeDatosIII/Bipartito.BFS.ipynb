```python
## Solana Navarro
## 906/22

##Importo Queue para usarlo en el BFS
from queue import Queue

## La idea principal es formar dos grupos de vertices y luego calcular cuantas aristas
## faltan para completar un grafo bipartito completo. Para esto, implemente un BFS
## que asigna a los hijos de un vertice el grupo opuesto al que pertenece su padre.

## Una vez definidos los grupos, procedi a calcular cuantas aristas serian necesarias
## para construir el grafo bipartito completo, conectando todos los vertices entre
## ambos grupos.

## Finalmente, reste este numero de aristas ideales menos la cantidad de aristas
## que ya existian en el grafo original, lo que me dio las aristas faltantes para
## completar el grafo bipartito.

def BFS(listadeadyacencias, v):
    cola = Queue()
    cola.put(v)
    grupos = [-1] * len(listadeadyacencias)
    grupos[v] = 1

    while(not cola.empty()):
      nodoactual = cola.get()
      for vecino in listadeadyacencias[nodoactual]:
          if grupos[vecino] == -1:
             if grupos[nodoactual] == 1:
                grupos[vecino] = 2
             if grupos[nodoactual] == 2:
                grupos[vecino] = 1
             cola.put(vecino)

    return grupos

##Input

n = int(input())
listadearistas = []
listadeadyacencias = [-1] * n
for i in range(n-1):
    arista = list(map(int, input().split()))
    listadearistas.append(tuple(arista))
    if listadeadyacencias[arista[0]-1] == -1:
        listadeadyacencias[arista[0]-1] = [arista[1]-1]
    else:
        listadeadyacencias[arista[0]-1].append(arista[1]-1)
    if listadeadyacencias[arista[1]-1] == -1:
        listadeadyacencias[arista[1]-1] = [arista[0]-1]
    else:
        listadeadyacencias[arista[1]-1].append(arista[0]-1)

grupos = BFS(listadeadyacencias, 0)
grupoA = 0
grupoB = 0
for j in range(len(grupos)):
   if grupos[j] == 1:
        grupoA += 1
   if grupos[j] == 2:
      grupoB += 1

print((grupoA*grupoB)-len(listadearistas))
```
