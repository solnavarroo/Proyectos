```python
import numpy as np

def BorW(numeros, cantidad, M, i, b, w):
  if i == cantidad:
     return(0)
  if M[i][b + 1][w + 1] == -1:

    opcion1 = 1 + BorW(numeros, cantidad, M, i+1, b, w)

    if numeros[i] > numeros[b] or b == -1:
       opcion2 = BorW(numeros, cantidad, M, i+1, i, w)
       opcion1 = min(opcion1, opcion2)
    if numeros[i] < numeros[w] or w == -1:
       opcion3 = BorW(numeros, cantidad, M, i+1, b, i)
       opcion1 = min(opcion1, opcion3)

    M[i][b + 1][w + 1] = opcion1


  return(M[i][b + 1][w + 1])


haysiguiente = True
res = []

while(haysiguiente):
  cantidad = int(input())

  if cantidad == -1:
    break

  numeros = list(map(int, input().split()))
  N = len(numeros)

  M = np.full((N + 1, N + 1, N), -1)

  res.append(BorW(numeros, N, M, 0, -1, -1))

for k in range(len(res)):
  print(res[k])


```
