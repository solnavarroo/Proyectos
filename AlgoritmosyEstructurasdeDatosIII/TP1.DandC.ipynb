```python
## Solana Navarro
## 906/22

## Resolví el problema de "Agressive Cows" utilizando el método de Divide and Conquer.
## La estrategia consiste en dividir repetidamente a la mitad las posibles distancias entre las vacas.
## Existen dos escenarios al evaluar una distancia posible:
##   - Si con la distancia propuesta es posible ubicar todas las vacas,
##     esto indica que esa distancia podría ser la mínima distancia de separación entre ellas.
##     Sin embargo, dado que hemos estado dividiendo y no hemos probado todas las opciones,
##     incrementamos la distancia en 1 y probamos con una distancia mayor.
##     Este proceso continúa hasta encontrar la mayor distancia que permite ubicar las C vacas.
##   - En el otro caso, si la distancia propuesta no permite ubicar todas las vacas,
##     significa que esa distancia es el límite superior y no tiene sentido probar distancias mayores.
##
## La distancia posible se calcula como el promedio entre la distancia máxima y mínima,
## es decir, es la mitad entre ambas.


## Funcion "esValidoponerCVacas" a partir de una distancia se fija cuantas vacas puede ubicar
## dependiendo de las posiciones de los stalls. Si pudo ubicar como minimo C vacas devuleve True.
def esValidoponerCVacas(stall, distancia, C):
    i = stall[0]
    j = 1
    vacaspuestas = 1
    while(j < len(stall)):

       if stall[j] >= i + distancia:
          i = stall[j]
          vacaspuestas += 1
       j += 1

    return(vacaspuestas >= C)

## Funcion "vacas" realiza lo explicado en la introduccion y devuelve el minimo
## de la maxima de todas las distancias posibles.
def vacas(stall, C):
    minimo = 0
    maximo = stall[-1] - stall[0]

    while(minimo < maximo):

       distancia = (maximo + minimo) // 2
       res = esValidoponerCVacas(stall, distancia, C)

       if res:
          minimo = distancia + 1
       else:
          maximo = distancia

    return(minimo-1)

## Input

T = int(input())
res = []

for i in range(T):
    [N, C] = list(map(int, input().split()))
    stall = []
    for j in range(N):
       xj = int(input())
       stall.append(xj)
    stall.sort()
    res.append(vacas(stall, C))

for k in range(len(res)):
    print(res[k])
```
